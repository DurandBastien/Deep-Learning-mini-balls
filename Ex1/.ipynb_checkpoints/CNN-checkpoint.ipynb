```python
import torch
import torchvision
import numpy as np
import matplotlib.pyplot as plt
from torch.nn import functional as F
import tp_src.dataset_det as d_loader
import torch.utils.data as torch_d
from PIL import Image
```

## The constants and the iterators ...:

```python
BATCHSIZE=10
dataset = d_loader.Balls_CF_Detection ("./train", 20999)
train_dataset, test_dataset = torch_d.random_split(dataset, [16799, 4200])
train_loader = torch.utils.data.DataLoader(train_dataset,batch_size=BATCHSIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset,batch_size=BATCHSIZE, shuffle=True)
COLORS = ['red', 'green', 'blue', 'yellow', 'lime', 'purple', 'orange', 'cyan', 'magenta']
STATS_INTERVAL = 50
```

## Have a glance at the data:

```python
def imshow(img):

    img = img / 2 + 0.5     # unnormalize    
    plt.imshow(img.permute(1, 2, 0).numpy().astype(np.uint8)) #convert pytorch tensor to np for printing image
    plt.show()

dataiter = iter(train_loader)
images, labels, pos = dataiter.next()
#     print(images[1].shape)
# print(labels)
for i in range(1):
    imshow(images[i])
    for item in np.nonzero(labels[i]):
        print(COLORS[item])
```

## CNN:

```python
def mapToBinary(x):
    a = x
    for i, prediction in enumerate(a):
        for j, ball_prediction in enumerate(prediction):
            if(ball_prediction < 0.5):
                a[i][j] = 0
            else:
                a[i][j] = 1
    return a
```

```python
class LeNet(torch.nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        self.conv1 = torch.nn.Conv2d(in_channels=3, out_channels = 6, kernel_size=5, groups=1)
        self.conv2 = torch.nn.Conv2d(6, 16, 5, 1)
        self.pool = torch.nn.MaxPool2d(4,4)
        self.fc1 = torch.nn.Linear(16*5*5, 100)
        self.fc2 = torch.nn.Linear(100, 9)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1,16*5*5)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        x = torch.sigmoid(x)
        #x = mapToBinary(x)
        return x
model = LeNet()
```

```python
from torchsummary import summary
summary(model,(3,100,100))
```

## Define loss function and optimizer:


```python
import torch.optim as optim

criterion = torch.nn.BCELoss()
#criterion = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

## Train the model!

```python
for epoch in range(2):
    running_loss = 0.0
    #get input 
    for batch_idx, (input, labels, bb) in enumerate(train_loader):
         # zero the parameter gradients
        optimizer.zero_grad()
        
        # forward + backward + optimize
        
        output = model(input)
        print(output)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()
        
        #print statistics
        running_loss += loss.item()
        if batch_idx % 200 == 199:    # print every 200 mini-batches
            print('[%d, %5d] loss: %.3f' %(epoch + 1, batch_idx + 1, running_loss / 200))
            running_loss = 0.0

print('Finished Training')
```

## Test network on the test data

```python
dataiter = iter(test_loader)
images, labels, pos= dataiter.next()
for i in range(10):
    imshow(images[i])
    for item in np.nonzero(labels[i]):
        print(COLORS[item])
```

## what does CNN think about this?

```python
print(images.shape)
outputs = model(images)
print(outputs.shape)
_, predicted = torch.topk(outputs.data,k = 3,dim = 1)
_, groundtruth = torch.topk(labels.data,k = 3, dim = 1)
correct = 0
for i in range(10):
    imshow(images[i])
    _,pos = torch.topk(outputs[i],3)
    for item in pos:
        print(COLORS[item])
        
    print("correct :",len(set(predicted[i].numpy()).intersection(set(groundtruth[i].numpy()))))
    if(len(set(predicted[i].numpy()).intersection(set(groundtruth[i].numpy()))) == 3):
        correct += 1
        
print()
print("Accuracy :", 100 * correct / labels.size(0))
```

```python
correct = 0
total = 0
with torch.no_grad():
    for data in test_loader:
        images, labels, bb= data
        outputs = model(images)
        _, predicted = torch.topk(outputs.data,k = 3,dim = 1)
        _, groundtruth = torch.topk(labels.data,k = 3, dim = 1)
        total += labels.size(0) * 3
        for i in range(predicted.size(0)):
            print(set(predicted[i].numpy()).intersection(set(groundtruth[i].numpy())))
            correct += len(set(predicted[i].numpy()).intersection(set(groundtruth[i].numpy())))

print('Accuracy of the network on the 10000 test images: %d %%' % (100 * correct / total))
```
